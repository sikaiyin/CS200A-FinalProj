<h4> Using the code for reference, take the given folder and create a dataframe with the picture object, and the encoding as listed below. </h4>	
0=Airplanes, 1=Bear, 2=Blimp, 3=Comet, 4=Crab, 5=Dog, 6=Dolphin, 7=Giraffe, 8=Goat, 9=Gorilla, 10=Kangaroo, 11=Killer-Whale, 12=Leopards, 13=Llama, 14= Penguin, 15= Porcupine, 16=Teddy-Bear, 17=Triceratops, 18=Unicorn, 19=Zebra

```python
encoding = {'airplanes':0, 'bear':1, 'blimp':2, 'comet':3, 'crab':4,\
            'dog':5, 'dolphin':6, 'giraffe':7, 'goat':8, 'gorilla':9,\
            'kangaroo':10, 'killer-whale':11, 'leopards':12, 'llama':13,\
            'penguin':14, 'porcupine':15, 'teddy-bear':16,'triceratops':17,\
            'unicorn':18, 'zebra':19}
def read_organize_data(file_path):
    img_df = pd.DataFrame(columns = ['Image','Picture','Encoding'])
    for dirpath, dirnames, filenames in os.walk(file_path):
        for file in filenames:
            img = io.imread(dirpath+'/'+file)
            row = {'Image':img,'Picture':file.split('_')[0],'Encoding':encoding[file.split('_')[0]]}
            img_df = img_df.append(row,ignore_index=True)
    return img_df

def create_validation_data(file_path):
    img_df = pd.DataFrame(columns = ['Image'])
    for dirpath, dirnames, filenames in os.walk(file_path):
        for file in filenames:
            img = io.imread(dirpath+'/'+file)
            row = {'Image':img}
            img_df = img_df.append(row,ignore_index=True)
    return img_df
    
    
```

```python
starting_data = read_organize_data("20_categories_training")
```

```python
validation_data = create_validation_data('20_Validation')
```
